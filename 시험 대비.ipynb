```python
# 왼쪽 피버팅
data = [5, 8, 2, 9, 4, 1, 3, 6]

def sol(start, end):
    if start >= end:
        return
    pivot = start
    left = start + 1
    right = end
    while left <= right:
        while left <= end and data[left]<=data[pivot]:  # left <= end 범위에 의한 전제 조건
            left = left + 1 #left += 1
        while right > start and data[right] >= data[pivot]:  # right > start 범위에 의한 전제 조건
            right  = right -1 #right -= 1
        if left > right: # j와 pivot 간의 자리 바꿈
            data[right], data[pivot] = data[pivot], data[right]
        else:
            data[left], data[right] = data[right], data[left]

    sol(start, right -1) # 작은 것들
    sol(right+1, end)

sol(0, 7)
print(data)
```

```python
# 오른쪽 피버팅
data = [5, 8, 2, 9, 4, 1, 3, 6]

def sol(start, end):
    global data
    if start >= end:
        return
    pivot = end
    left = start
    right = end - 1
    while left <= right:
        while left < end and data[left] <= data[pivot]:
            left += 1
        while right >= start and data[right] >= data[pivot]:
            right -= 1
        if left > right:
            data[left], data[pivot] = data[pivot], data[left]
        else:
            data[right], data[left] = data[left], data[right]
    sol(start, left - 1)
    sol(left+1, end)
sol(0, 7)
data
```

```python
# 최소 점프 횟수 프로그램
'''
한 번에 이동할 수 있는 점프 거리가 여러 종류 있다.
예를 들어 1칸, 3칸, 5칸 점프가 가능하다고 하자.
목표 지점 N에 도착하기 위해 필요한 최소 점프 횟수를 구하는 프로그램을 작성하시오.
'''

# 분기 한정 법
moves = [1, 3, 5]

N = 10


final = 2**9999999

def sol(k=[]):
    global moves, final, N
    if sum(k) > N:
        return


    if sum(k) == N:
        #print(k)
        if len(k) < final: final = len(k)
        return
    
    for i in moves:
        k.append(i)
        sol(k)
        k.pop()

sol()
print(final)
```

```python
# 최소 점프 횟수 프로그램
'''
한 번에 이동할 수 있는 점프 거리가 여러 종류 있다.
예를 들어 1칸, 3칸, 5칸 점프가 가능하다고 하자.
목표 지점 N에 도착하기 위해 필요한 최소 점프 횟수를 구하는 프로그램을 작성하시오.
'''

# 탐욕법
moves = [1, 3, 5]

N = 10

ans = 0

c = 0

while ans < N:
    c += 1
    possible = []
    for i in moves:
        if (N - ans) >= i:
            possible.append(i)
    if possible == []:
        break
    ans += max(possible)
    if ans == N:
        break

print(c)
```

```python
# 최소 이동 시간 계산 프로그램
'''
여러 개의 섬이 다리로 연결되어 있다.
각 다리는 정확한 이동 시간(비용) 이 있으며, 
일부 섬은 직접 연결되어 있지 않다.
두 섬 A와 B가 주어졌을 때,
A에서 B까지 이동하는 데 걸리는 
최소 이동 시간을 
구하는 프로그램을 작성하시오.
'''

'''
섬 개수 N = 6
다리 정보(양방향):
1 2 7
1 3 9
2 4 10
3 5 2
4 6 1
5 6 5
출발 = 1, 도착 = 6
'''

N = 6

start = 1

end = 6

edges = [
    (1, 2, 7),
    (1, 3, 9),
    (2, 4, 10),
    (3, 5, 2),
    (4, 6, 1),
    (5, 6, 5)
]

M = [[0] * N for i in range(N)]
for a, b, w in edges:
    print(a, b, w)
    M[a - 1][b - 1] = w
    M[b - 1][a - 1] = w

#print(M)

ans = 1278346
final = []

def sol(x, v, k, w): # x는 현재 v는 지금 까지 노드 수, k는 노드 리스트, w는 시간 합산
    #print(k)
    global M, final, N, ans
    if 0 in k and 5 in k:
        if w < ans:
            ans = w
            final = k.copy()
        return        
    
    for i in range(len(M)):
        if i not in k and M[x][i] != 0:
            k.append(i)
            sol(i, v+1, k, w + M[x][i])
            k.pop()

sol(0, 1, [0], 0)
print(ans)
print(final)
```

```python
import graphviz

# 다이어그램 생성
dot = graphviz.Digraph('RL_Architecture', comment='RL Agent Architecture')
dot.attr(rankdir='LR', size='12,8', dpi='300')

# 노드 기본 스타일 설정
dot.attr('node', shape='box', style='rounded,filled', fontname='Sans')

# 1. Policy Module (정책) - 파란색 계열
with dot.subgraph(name='cluster_0') as c:
    c.attr(label='1. Policy Module (두뇌)', style='filled', color='#E1F5FE')
    c.node('Prob_Table', '명령어 산출\n(확률표 & 가중치)', fillcolor='white')
    c.node('Param_Sel', '파라미터 선정\n(호환성 체크)', fillcolor='white')
    c.node('Epsilon', '엡실론/가우시안 필터\n(탐험/자유도 부여)', fillcolor='#B3E5FC')
    c.node('Final_Cmd', '최종 명령어 생성\n(JSON)', shape='parallelogram', fillcolor='#81D4FA')
    
    c.edge('Prob_Table', 'Param_Sel')
    c.edge('Param_Sel', 'Epsilon')
    c.edge('Epsilon', 'Final_Cmd')

# 2. DMP / Action Module (실행) - 주황색 계열
with dot.subgraph(name='cluster_1') as c:
    c.attr(label='2. DMP / Action (실행)', style='filled', color='#FFF3E0')
    c.node('Exec', 'Action 실행\n(subprocess / nmap)', shape='component', fillcolor='#FFCC80')
    c.node('Parser', '로그 파싱 & 필터링', fillcolor='white')
    c.node('State_Check', '상태(State) 업데이트', fillcolor='white')
    c.node('Err_Check', '오류 판정\n(Return Code)', fillcolor='#FFAB91')
    
    c.edge('Exec', 'Parser')
    c.edge('Parser', 'State_Check')
    c.edge('Parser', 'Err_Check')

# 3. Reward Module (보상) - 초록색 계열
with dot.subgraph(name='cluster_2') as c:
    c.attr(label='3. Reward Module (평가)', style='filled', color='#E8F5E9')
    c.node('Flag_R', 'FLAG (목표달성)', fillcolor='#A5D6A7')
    c.node('Error_R', 'ERROR (페널티)', fillcolor='#EF9A9A')
    c.node('Curiosity_R', 'CURIOSITY (호기심)', fillcolor='#FFF59D')
    c.node('Prophecy_R', 'PROPHECY (예측점수)', fillcolor='#CE93D8')
    c.node('Total_Reward', '최종 보상 (r_tot)\nTanh(Sum)', shape='doublecircle', fillcolor='#66BB6A', fontcolor='white')
    
    c.edge('Flag_R', 'Total_Reward')
    c.edge('Error_R', 'Total_Reward')
    c.edge('Curiosity_R', 'Total_Reward')
    c.edge('Prophecy_R', 'Total_Reward')

# 4. Prophecy / Memory (학습) - 보라색 계열
with dot.subgraph(name='cluster_3') as c:
    c.attr(label='4. Prophecy / Memory (기억)', style='filled', color='#F3E5F5')
    c.node('History', 'History 저장\n(State, Action, Reward)', shape='folder', fillcolor='white')
    c.node('LSTM', 'LSTM 모델\n(시계열 학습)', fillcolor='#E1BEE7')
    c.node('Imagination', '상상(Imagination)\n다음 상태 예측', fillcolor='white')
    
    c.edge('History', 'LSTM')
    c.edge('LSTM', 'Imagination')

# 전체 흐름 연결 (모듈 간 연결)
dot.edge('Final_Cmd', 'Exec', label='명령어 입력')
dot.edge('Exec', 'Flag_R', label='Flag 획득?')
dot.edge('Err_Check', 'Error_R', label='에러 발생?')
dot.edge('Parser', 'Curiosity_R', label='새 지식?')

# 학습 루프 (Feedback Loop)
dot.edge('Total_Reward', 'Prob_Table', label='가중치 업데이트 (Learning)', color='blue', penwidth='2.0')
dot.edge('State_Check', 'Prob_Table', label='현재 상태 (State)', style='dashed')
dot.edge('State_Check', 'History', label='경험 축적')
dot.edge('Imagination', 'Prophecy_R', label='예측 vs 실제 비교')

# 다이어그램 출력
print(dot.source)
```
