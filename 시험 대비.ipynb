```python
'''í€µ'''
a = [5, 4, 3, 2, 1]

def sol(start, end):
    global a
    if start >= end:
        return
    
    pivot = end
    left = start
    right = end - 1

    while left <= right:
        while left < end and a[left] <= a[pivot]: # pivotì„ ì œì™¸í•œ ë¦¬ìŠ¤íŠ¸ ì•ˆì—ì„œë§Œ ë…¸ë‹ˆê¹Œ
            left += 1
        while right >= start and a[right] >= a[pivot]:
            right -= 1

        if left>right:
            a[left], a[pivot] = a[pivot], a[left]
        else:
            a[left], a[right] = a[right], a[left]
    sol(start, left-1)
    sol(left+1,end)

sol(0, len(a)-1)
print(a)
```

```python
# ìµœì†Œ ì í”„ íšŸìˆ˜ í”„ë¡œê·¸ë¨
'''
í•œ ë²ˆì— ì´ë™í•  ìˆ˜ ìˆëŠ” ì í”„ ê±°ë¦¬ê°€ ì—¬ëŸ¬ ì¢…ë¥˜ ìˆë‹¤.
ì˜ˆë¥¼ ë“¤ì–´ 1ì¹¸, 3ì¹¸, 5ì¹¸ ì í”„ê°€ ê°€ëŠ¥í•˜ë‹¤ê³  í•˜ì.
ëª©í‘œ ì§€ì  Nì— ë„ì°©í•˜ê¸° ìœ„í•´ í•„ìš”í•œ ìµœì†Œ ì í”„ íšŸìˆ˜ë¥¼ êµ¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì‹œì˜¤.
'''

# íƒìš•ë²•
moves = [1, 3, 5]

N = 10

ans = 0

c = 0

while ans < N:
    c += 1
    possible = []
    for i in moves:
        if (N - ans) >= i:
            possible.append(i)
    if possible == []:
        break
    ans += max(possible)
    if ans == N:
        break

print(c)
```
