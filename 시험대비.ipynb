```python
# 최소 점프 횟수 프로그램
'''
한 번에 이동할 수 있는 점프 거리가 여러 종류 있다.
예를 들어 1칸, 3칸, 5칸 점프가 가능하다고 하자.
목표 지점 N에 도착하기 위해 필요한 최소 점프 횟수를 구하는 프로그램을 작성하시오.
'''

# 분기 한정 법
moves = [1, 3, 5]

N = 10


final = 2**9999999

def sol(k=[]):
    global moves, final, N
    if sum(k) > N:
        return


    if sum(k) == N:
        print(k)
        if len(k) < final: final = len(k)
        return
    
    for i in moves:
        k.append(i)
        sol(k)
        k.pop()

sol()
print(final)
```

```python
# 최소 점프 횟수 프로그램
'''
한 번에 이동할 수 있는 점프 거리가 여러 종류 있다.
예를 들어 1칸, 3칸, 5칸 점프가 가능하다고 하자.
목표 지점 N에 도착하기 위해 필요한 최소 점프 횟수를 구하는 프로그램을 작성하시오.
'''

# 탐욕법
moves = [1, 3, 5]

N = 10

ans = 0

c = 0

while ans < N:
    c += 1
    possible = []
    for i in moves:
        if (N - ans) >= i:
            possible.append(i)
    if possible == []:
        break
    ans += max(possible)
    if ans == N:
        break

print(c)
```

```python
# 최소 이동 시간 계산 프로그램
'''
여러 개의 섬이 다리로 연결되어 있다.
각 다리는 정확한 이동 시간(비용) 이 있으며, 
일부 섬은 직접 연결되어 있지 않다.
두 섬 A와 B가 주어졌을 때,
A에서 B까지 이동하는 데 걸리는 
최소 이동 시간을 
구하는 프로그램을 작성하시오.
'''

'''
섬 개수 N = 6
다리 정보(양방향):
1 2 7
1 3 9
2 4 10
3 5 2
4 6 1
5 6 5
출발 = 1, 도착 = 6
'''

N = 6

start = 1

end = 6

edges = [
    (1, 2, 7),
    (1, 3, 9),
    (2, 4, 10),
    (3, 5, 2),
    (4, 6, 1),
    (5, 6, 5)
]

M = [[0] * N for i in range(N)]
for a, b, w in edges:
    print(a, b, w)
    M[a - 1][b - 1] = w
    M[b - 1][a - 1] = w

print(M)

ans = 1278346
final = []

def sol(x, v, k, w): # x는 현재 v는 지금 까지 노드 수, k는 노드 리스트, w는 시간 합산
    print(k)
    global M, final, N, ans
    if 0 in k and 5 in k:
        if w < ans and k[-1] == 5:
            ans = w
            final = k
        return        
    
    for i in range(len(M)):
        if i not in k and M[x][i] != 0:
            k.append(i)
            sol(i, v+1, k, w + M[x][i])
            k.pop()

sol(0, 1, [0], 0)
print(ans)
print(final)
```
